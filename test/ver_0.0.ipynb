```python
import cv2
from ultralytics import YOLO
import numpy as np

# Load YOLO model
model = YOLO('yolov8n.pt')  # Ensure the model file is accessible

# Load video file
video_path = '(334) [에필로그] 김지원 거짓말에 n년째 속고 있는 순박 청년 김수현ㅋㅋㅋ #눈물의여왕 EP.7 _ tvN 240330 방송 - YouTube 외 페이지 2개 - 개인 - Microsoft​ Edge 2024-03-31 12-42-54.mp4' 
cap = cv2.VideoCapture(video_path)
coor=np.zeros([4,10])
vars=[]
count=0

# Loop through video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    # Perform prediction
    results = model(frame[:, :, ::-1])
    plots = results[0].plot()
    boxes = results[0].boxes

    for box in boxes :
        if box.cls.cpu().detach().numpy().tolist()[0] != 0 : ## 사람이 0으로 레이블 됨.
            continue
        
        x1, y1, x2, y2 = box.xyxy.cpu().detach().numpy().tolist()[0]
        if count==0:
            coor = np.array([[x1, y1, x2, y2] for _ in range(10)]).T
            var = np.var(coor[0])/np.mean(coor[0])+np.var(coor[1])/np.mean(coor[1])+np.var(coor[2])/np.mean(coor[2])+np.var(coor[3])/np.mean(coor[3])
            vars.append(var)
            print(vars)
        elif count%5 == 0:
            coor[:, (count//5)%10] = [x1, y1, x2, y2]
            var = np.var(coor[0])/np.mean(coor[0])+np.var(coor[1])/np.mean(coor[1])+np.var(coor[2])/np.mean(coor[2])+np.var(coor[3])/np.mean(coor[3])
            vars.append(var)
            if var > 1:
                print("변했다!")
                plots=cv2.cvtColor(plots, cv2.COLOR_BGR2RGB)
                cv2.imshow("plot", plots)
                coor = np.array([[x1, y1, x2, y2] for _ in range(10)]).T
            print(vars)
            print(coor)
            break
        
    count+=1
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Break the loop if 'q' is pressed
        break
# Release resources
cap.release()
cv2.destroyAllWindows()
```
