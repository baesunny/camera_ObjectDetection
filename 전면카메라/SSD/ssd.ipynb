```python
import torch
import torchvision
import cv2
```

```python
model = torchvision.models.detection.ssd300_vgg16(pretrained=True)
```

```python
model.eval()
```

```python
# Check if CUDA is available and set the device accordingly
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")
model.to(device)
```

```python
image = cv2.imread('Hero.png')
print(type(image))
img = image.copy()
```

```python
from torchvision import transforms as T
imgtransform = T.ToTensor()
image = imgtransform(image)
print(type(image))
```

```python
with torch.no_grad():
    prediction = model([image])
    print(prediction[0].keys())

    bbox, scores, labels = prediction[0]['boxes'], prediction[0]['scores'],prediction[0]['labels']

    nums = torch.argwhere(scores>0.50).shape[0]
    for i in range(nums):
        x,y,w,h = bbox[i].numpy().astype('int')
        cv2.rectangle(img,(x,y),(w,h),(0,0,255),5)

cv2.imshow('frame',img)
cv2.waitKey(0)
```

# 여기서부터는 실시간 cam

```python
# camera setting 
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

# constants
GREEN = (0, 255, 0)       
WHITE = (255, 255, 255)   
RED = (0, 0, 255)
BLUE = (255,0,0)

import time
while True:
    start = time.time()
    success, frame = cap.read()
    if not success:
        print('Cam Error')
        break

    
    with torch.no_grad():
         frame_tensor = imgtransform(frame).to(device)
         pred = model([frame_tensor])

         bbox, scores, labels = pred[0]['boxes'], pred[0]['scores'],pred[0]['labels']
         nums = torch.argwhere(scores>0.50).shape[0]
         for i in range(nums):
            x,y,w,h = bbox[i].numpy().astype('int')
            label = labels[i].numpy().astype('int')
            cv2.rectangle(frame,(x,y),(w,h),BLUE,2)
            cv2.putText(frame, f"{label}", (x,y-10),cv2.FONT_ITALIC,1,WHITE,2)

    
    end = time.time()
    totaltime = end-start
    fps = f'FPS: {1 / totaltime:.2f}'
    print(fps)

    cv2.putText(frame, fps, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, BLUE, 2)
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) == ord('q'):
            break
    
cap.release()
cv2.destroyAllWindows()
```
